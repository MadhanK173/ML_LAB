```python
# ==============================
# STEP 1: Upload CSV / Excel file
# ==============================
from google.colab import files
uploaded = files.upload()

# ==============================
# STEP 2: Import libraries
# ==============================
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# ==============================
# STEP 3: Load dataset
# ==============================
# Change file name if needed
try:
    data = pd.read_csv("data.csv")   # if CSV
except:
    data = pd.read_excel("iris (2) (1).xlsx")   # if Excel

print("Dataset Loaded:\n")
print(data.head())

# ==============================
# STEP 4: Use only numeric columns
# ==============================
X = data.select_dtypes(include=['int64', 'float64'])

# ==============================
# STEP 5: Choose number of clusters (k)
# ==============================
k = 3   # You can change this

# ==============================
# STEP 6: Apply K-Means
# ==============================
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)

# ==============================
# STEP 7: Get cluster labels
# ==============================
labels = kmeans.labels_
data['Cluster'] = labels

print("\nClustered Data:\n")
print(data.head())

# ==============================
# STEP 8: Display Centroids
# ==============================
print("\nCentroids:\n")
print(kmeans.cluster_centers_)

# ==============================
# STEP 9: Visualization (first 2 features)
# ==============================
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=labels)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K-Means Clustering")
plt.show()
```
