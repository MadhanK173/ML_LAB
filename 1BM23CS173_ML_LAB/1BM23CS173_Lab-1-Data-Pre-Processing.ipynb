```python
# ==============================
# DIABETES DATASET (YOUR FILE)
# ==============================

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler

# Upload file in Colab
from google.colab import files
uploaded = files.upload()

# Load dataset (use exact file name after upload)
df = pd.read_csv('Dataset of Diabetes .csv')

print("Original Data:\n", df.head())

# ------------------------------
# 1. DATA CLEANING
# ------------------------------

# Replace invalid zeros with NaN (common in diabetes dataset)
# Identified relevant numerical columns from the actual DataFrame
cols = ['Urea', 'Cr', 'HbA1c', 'Chol', 'TG', 'HDL', 'LDL', 'VLDL', 'BMI']
df[cols] = df[cols].replace(0, np.nan)

# Fill missing values with median
df.fillna(df.median(numeric_only=True), inplace=True)

# ------------------------------
# OUTLIER HANDLING (IQR)
# ------------------------------
Q1 = df.quantile(0.25, numeric_only=True)
Q3 = df.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1

# Select only numeric columns for outlier detection comparison
numeric_df = df.select_dtypes(include=np.number)
outlier_mask = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)
df_clean = df[~outlier_mask].copy()

# Handle categorical data after cleaning and before scaling
le = LabelEncoder()
categorical_cols = df_clean.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df_clean[col] = le.fit_transform(df_clean[col])

print("\nAfter Cleaning and Categorical Encoding:\n", df_clean.shape)

# ------------------------------
# 2. DATA TRANSFORMATION
# ------------------------------

# Min-Max Scaling
minmax = MinMaxScaler()
df_minmax = pd.DataFrame(minmax.fit_transform(df_clean), columns=df_clean.columns)

# Standard Scaling
scaler = StandardScaler()
df_standard = pd.DataFrame(scaler.fit_transform(df_clean), columns=df_clean.columns)

print("\nMin-Max Scaled:\n", df_minmax.head())
print("\nStandard Scaled:\n", df_standard.head())
```

```python
# ==============================
# ADULT DATASET (YOUR FILE)
# ==============================

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler

# Upload file
from google.colab import files
uploaded = files.upload()

# Load dataset
df = pd.read_csv('adult.csv')

print("Original Data:\n", df.head())

# ------------------------------
# 1. DATA CLEANING
# ------------------------------

# Replace '?' with NaN
df.replace('?', np.nan, inplace=True)

# Drop missing values
df.dropna(inplace=True)

# ------------------------------
# HANDLE CATEGORICAL DATA
# ------------------------------

le = LabelEncoder()
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# ------------------------------
# OUTLIER HANDLING (IQR)
# ------------------------------
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

df_clean = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

print("\nAfter Cleaning:\n", df_clean.shape)

# ------------------------------
# 2. DATA TRANSFORMATION
# ------------------------------

# Min-Max Scaling
minmax = MinMaxScaler()
df_minmax = pd.DataFrame(minmax.fit_transform(df_clean), columns=df_clean.columns)

# Standard Scaling
scaler = StandardScaler()
df_standard = pd.DataFrame(scaler.fit_transform(df_clean), columns=df_clean.columns)

print("\nMin-Max Scaled:\n", df_minmax.head())
print("\nStandard Scaled:\n", df_standard.head())
```
